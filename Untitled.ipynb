```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# 1. Load data (robust handling)
df = pd.read_csv('imdb_labelled.txt', sep='\t', header=None, names=['text', 'label'])
df = df.dropna(subset=['label'])          # drop rows where label is missing
df['label'] = df['label'].astype(int)     # ensure integer labels
print(f'Dataset size: {df.shape[0]}')

# 2. Train/validation split (80/20, stratified)
X_train, X_val, y_train, y_val = train_test_split(
    df['text'], df['label'],
    test_size=0.2,
    random_state=42,
    stratify=df['label']
)

# 3. TF-IDF vectorisation (fit only on training data)
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_val_tfidf = vectorizer.transform(X_val)

# 4. Logistic Regression model
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_tfidf, y_train)

# 5. Evaluation
y_pred = model.predict(X_val_tfidf)
print(f'Validation Accuracy: {accuracy_score(y_val, y_pred):.4f}\n')
print('Classification Report:')
print(classification_report(y_val, y_pred, target_names=['Negative', 'Positive']))
print('Confusion Matrix:')
print(confusion_matrix(y_val, y_pred))

# 6. Predict on new sentences (optional demo)
new_texts = [
    "This movie was absolutely wonderful and I loved every minute of it.",
    "A complete waste of time, the plot was nonexistent.",
    "The acting was decent but the story was a bit boring.",
    "Brilliant masterpiece, highly recommended!"
]
new_tfidf = vectorizer.transform(new_texts)
preds = model.predict(new_tfidf)
print('\nPredictions on new sentences:')
for text, pred in zip(new_texts, preds):
    print(f'{text:70} -> {"Positive" if pred==1 else "Negative"}')
```
